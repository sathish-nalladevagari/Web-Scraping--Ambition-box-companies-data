```python
#necessary imports

from bs4 import BeautifulSoup
import requests
```

```python
webpage = requests.get("https://www.ambitionbox.com/list-of-companies?page=1").text
```

```python
soup = BeautifulSoup(webpage,'lxml')
```

```python
companies = soup.find_all('div',class_="company-content-wrapper")
```

```python
c_name = []
place = []
c_type = []
old = []
emp = []
rating = []
review = []




for company in companies:
  c_name.append(company.find('h2',class_="company-name bold-title-l").text.strip())
  place.append(company.find_all('p',class_="infoEntity sbold-list-header")[1].get_text(strip=True))
  c_type.append(company.find_all('p',class_="infoEntity sbold-list-header")[0].get_text(strip=True))
  old.append(company.find_all('p',class_="infoEntity sbold-list-header")[2].get_text(strip=True))
  emp.append(company.find_all('p',class_="infoEntity sbold-list-header")[3].get_text(strip=True))
  rating.append(company.find('p',class_='rating').text.strip())
  review.append(company.find('a',class_='review-count').text.strip())
  

data = {'Company':c_name,'rating':rating,'reviews':review,'place':place,"company type":c_type,'old':old,'employees':emp}

  



  
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(data)
```
